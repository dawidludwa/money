```python
df = df[['Data transakcji', 'Dane kontrahenta', 'Tytuł',
       'Nr rachunku', 'Szczegóły', 'Nr transakcji',
       'Kwota transakcji (waluta rachunku)']]
```

```python
columns = {
    'Data transakcji': 'date',
    'Dane kontrahenta': 'contractor',
    'Tytuł': 'title',
    'Szczegóły': 'details',
    'Nr rachunku': 'account',
    'Nr transakcji': 'transaction',
    'Kwota transakcji (waluta rachunku)': 'amount',
}
df.rename(columns = columns, inplace = True)
```

```python
f"{df['date'].min()} {df['date'].max()}"
```

```python
df.head()
```

```python
df.info()
```

```python
# remove all rows if value in column is NaN
df.dropna(subset=['amount'], inplace=True)

# remove whitespace from sting columns
df['title'] = df['title'].str.strip()
df['contractor'] = df['contractor'].str.strip()
df['account'] = df['account'].str.strip()
df['transaction'] = df['transaction'].str.strip()

# remove qoutes from string columns
df['account'] = df['account'].str.replace('"', '')
df['transaction'] = df['transaction'].str.replace('"', '')
df['account'] = df['account'].str.replace("'", '')
df['transaction'] = df['transaction'].str.replace("'", '')
```

```python
df = df[['date', 'contractor', 'title', 'account', 'transaction', 'amount']]
```

```python
df[df['title'] == 'Przelew własny']
own_transfers = df[df['title'] == 'Przelew własny']
df = df[df['title'] != 'Przelew własny']
```
